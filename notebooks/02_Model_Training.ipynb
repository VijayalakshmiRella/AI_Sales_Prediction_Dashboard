```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

import joblib
```

```python
df = pd.read_csv(
    r"D:\Ai_project\AI_Sales_Prediction_Dashboard\processed_data\cleaned_data.csv"
)
```

```python
features = [
    "year",
    "month",
    "quantity",
    "discount",
    "profit",
    "category",
    "region",
    "segment"
]

X = df[features]

y = df["sales"]
```

```python
X = pd.get_dummies(
    X,
    drop_first=True
)
```

```python
X.head()
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
print(X_train.shape)
print(X_test.shape)

print(y_train.shape)
print(y_test.shape)
```

```python
linear_model = LinearRegression()

linear_model.fit(
    X_train,
    y_train
)
```

```python
linear_predictions = linear_model.predict(
    X_test
)
```

```python
mae = mean_absolute_error(
    y_test,
    linear_predictions
)

print("MAE:", mae)
```

```python
mse = mean_squared_error(
    y_test,
    linear_predictions
)

print("MSE:", mse)
```

```python
rmse = np.sqrt(mse)

print("RMSE:", rmse)
```

```python
r2 = r2_score(
    y_test,
    linear_predictions
)

print("R² Score:", r2)
```

```python
comparison = pd.DataFrame({

    "Actual Sales": y_test.values,

    "Predicted Sales": linear_predictions

})

comparison.head(10)
```

```python
rf_model = RandomForestRegressor(

    n_estimators=100,

    random_state=42

)

rf_model.fit(

    X_train,

    y_train

)
```

```python
rf_predictions = rf_model.predict(

    X_test

)
```

```python
print(

    "MAE:",

    mean_absolute_error(

        y_test,

        rf_predictions

    )

)

print(

    "RMSE:",

    np.sqrt(

        mean_squared_error(

            y_test,

            rf_predictions

        )

    )

)

print(

    "R²:",

    r2_score(

        y_test,

        rf_predictions

    )

)
```
